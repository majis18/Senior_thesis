```python
import pandas as pd
```

```python
path = './brca_metabric/'
fname_data_mrna_all = 'data_mrna_agilent_microarray_zscores_ref_all_samples.txt'
fname_data_patient = 'data_clinical_patient.txt'
```

```python
df_mrna = pd.read_table(path+fname_data_mrna_all)
df_mrna = df_mrna.drop('Entrez_Gene_Id', axis=1)
df_mrna = df_mrna.rename(columns={'Hugo_Symbol': 'PATIENT_ID'})
df_mrna = df_mrna.set_index('PATIENT_ID')
df_mrna = df_mrna.T
df_mrna.columns.name = None
df_mrna.index.name = None
df_mrna = df_mrna.reset_index()
df_mrna = df_mrna.rename(columns={'index': 'PATIENT_ID'})
df_mrna
```

```python
df_patient = pd.read_table(path+fname_data_patient, header = 4)
df_patient
```

```python
df_fill_NaN_mean = df_mrna.fillna(df_mrna.mean())
print(df_fill_NaN_mean.isnull().values.sum())
```

```python
df_rfs = df_patient[['PATIENT_ID','RFS_STATUS','RFS_MONTHS']]
df_rfs.isnull().sum()
```

```python
df_null_col = df_rfs.loc[:, df_rfs.isnull().any()]
df_null = df_null_col[df_null_col.isnull().any(axis=1)]
df_null.to_csv('null_data.csv')
```

```python
df_rfs.query('RFS_STATUS == "0:Not Recurred"')
```

```python
df_master = pd.merge(df_rfs, df_fill_NaN_mean, on='PATIENT_ID', how='right')
df_master
```

```python
df_master[['PATIENT_ID','RFS_STATUS','RFS_MONTHS']].isnull().sum()
```

```python
df_master = df_master.dropna()
df_master = df_master.drop('PATIENT_ID', axis=1)
df_master = pd.get_dummies(df_master, drop_first=True)
df_master
```

```python
rfs_status_col = df_master.pop('RFS_STATUS_1:Recurred')
df_master.insert(1,'RFS_STATUS_1:Recurred',rfs_status_col)
df_master
```

```python
df_master['RFS_STATUS_1:Recurred'].value_counts()
```

```python
df_master1 = df_master.drop('RFS_STATUS_1:Recurred',axis=1).copy()
```
