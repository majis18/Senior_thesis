```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df0 = pd.read_csv('/Users/shota_majima_tus/Desktop/ゼミ4/history_data/test/1_test_acc_log.csv')
df1 = pd.read_csv('/Users/shota_majima_tus/Desktop/ゼミ4/history_data/test/1_test_loss_log.csv')
```

```python
df = pd.concat([df0, df1], axis=0).drop('Unnamed: 0', axis=1)
df = df.set_axis(['acc', 'loss']).T
li1 = df['loss'].to_list()
li2 = df['acc'].to_list()
plt.plot(range(len(li1)), li1, color='Blue')
plt.ylabel('Loss')
plt.xlabel('epoch')
plt.title('Train Loss')
plt.savefig('Test_train_loss.png')
plt.show()
```

```python
plt.plot(range(len(li2)), li2, color='Red')
plt.ylabel('Accuracy')
plt.xlabel('epoch')
plt.title('Train Accuracy')
plt.savefig('Test_train_acc.png')
plt.show()
```
